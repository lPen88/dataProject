```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

df = pd.read_csv('dataset/dataset_l_fix.csv')

df = df[df['Sleep Hours Per Day'] != 0]
df.loc[:, 'Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})

Y = df['Heart Attack Risk']
X=df[['Age', 'Sex', 'Smoking', 'Alcohol Consumption', 'Cholesterol', 'Systolic', 'Diastolic', 'Heart Rate']]

correlation_matrix = X.corr()

plt.figure(figsize=(12, 10))  # Increase the figure size
import seaborn as sns
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".3f",            # Show only 2 decimal places
    cmap='coolwarm',
    square=True,
    annot_kws={'size': 8} # Smaller font for annotations
)
plt.title('Correlation Matrix')
plt.show()
```
